```python
plt.figure(figsize=(8,5))
plt.hist(df["G3"], bins=10, color="skyblue", edgecolor="black")

plt.title("Distribusi Nilai Akhir Siswa")
plt.xlabel("Nilai G3")
plt.ylabel("Jumlah Siswa")

plt.show()
```

plt.hist() digunakan untuk membuat histogram. df["G3"] adalah data nilai akhir yang divisualisasikan. bins=10 menentukan jumlah interval pada histogram. color menentukan warna batang. edgecolor memberi warna garis tepi batang.

Histogram digunakan untuk melihat sebaran nilai akhir siswa (G3). Dilihat dari tinggi batang pada grafik, dapat diketahui nilai yang paling banyak diperoleh siswa serta bagaimana distribusi nilai dalam dataset.

```python
plt.figure(figsize=(8,5))

x = ["G1", "G2", "G3"]
y = [df["G1"].mean(), df["G2"].mean(), df["G3"].mean()]

plt.plot(x, y, 'mo-')

plt.xlabel('Tahap Penilaian')
plt.ylabel('Rata-rata Nilai')
plt.title('Perkembangan Rata-rata Nilai Siswa')

plt.show()
```

dihitung rata-rata nilai dari kolom G1, G2, dan G3 menggunakan fungsi mean(). Variabel x berisi nama tahap penilaian, sedangkan variabel y berisi nilai rata-rata dari setiap tahap. Fungsi plt.plot(x, y, 'mo-') digunakan untuk membuat diagram garis dengan warna magenta (m), marker lingkaran (o), dan garis penghubung (-). Label sumbu dan judul grafik ditambahkan menggunakan plt.xlabel(), plt.ylabel(), dan plt.title().



Dilihat dari grafik, visualisasi ini menunjukkan perubahan rata-rata nilai siswa dari penilaian pertama (G1), penilaian kedua (G2), hingga nilai akhir (G3). Dengan diagram garis ini dapat terlihat apakah nilai siswa cenderung meningkat, menurun, atau tetap stabil pada setiap tahap penilaian. Grafik ini membantu memahami tren performa akademik siswa secara keseluruhan.

```python
plt.figure(figsize=(8,5))
plt.scatter(df["absences"], df["G3"], color="red")

plt.title("Hubungan Absensi dengan Nilai")
plt.xlabel("Jumlah Absensi")
plt.ylabel("Nilai G3")

plt.show()
```

plt.scatter() digunakan untuk membuat scatter plot. df["absences"] digunakan sebagai sumbu x. df["G3"] digunakan sebagai sumbu y. color menentukan warna titik pada grafik.

Scatter plot digunakan untuk melihat hubungan antara jumlah absensi dan nilai akhir siswa. Dilihat dari penyebaran titik pada grafik, dapat diamati apakah siswa dengan absensi lebih banyak cenderung memiliki nilai yang lebih rendah.

```python
x = df["absences"]
g1 = df["G1"]
g2 = df["G2"]
g3 = df["G3"]

plt.figure(figsize=(8,5))

# scatter G1
plt.scatter(x, g1, color="green", marker="o", label="G1")

# scatter G2
plt.scatter(x, g2, color="blue", marker="s", label="G2")

# scatter G3
plt.scatter(x, g3, color="red", marker="^", label="G3")

plt.title("Perbandingan Absensi dan Nilai Siswa")
plt.xlabel("Jumlah Absensi")
plt.ylabel("Nilai")

plt.legend()
plt.show()
```

diambil kolom absences, G1, G2, dan G3. Fungsi plt.scatter() digunakan untuk menampilkan ketiga variabel nilai dalam bentuk titik. Parameter marker digunakan untuk membedakan bentuk titik, yaitu lingkaran (o) untuk G1, kotak (s) untuk G2, dan segitiga (^) untuk G3. Warna juga diatur berbeda agar grafik lebih mudah dibaca. Fungsi plt.legend() digunakan untuk menampilkan keterangan setiap variabel.

Visualisasi ini menampilkan hubungan antara jumlah absensi siswa dengan nilai G1, G2, dan G3 dalam bentuk titik. Dilihat dari grafik, setiap nilai ditampilkan dengan warna dan bentuk marker yang berbeda sehingga memudahkan perbandingan antar variabel. Dengan visualisasi ini dapat diamati apakah terdapat pola tertentu antara jumlah absensi siswa dengan nilai yang diperoleh.

```python
plt.figure(figsize=(7,5))
sns.boxplot(x="studytime", y="G3", data=df, palette="Set2")

plt.title("Perbandingan Nilai Berdasarkan Study Time")

plt.show()
```

sns.boxplot() digunakan untuk membuat boxplot menggunakan library Seaborn. x="studytime" menunjukkan kategori waktu belajar. y="G3" menunjukkan nilai akhir siswa. data=df menunjukkan dataset yang digunakan. palette menentukan warna grafik.

Boxplot digunakan untuk membandingkan nilai siswa berdasarkan waktu belajar. Dilihat dari median dan sebaran data pada setiap kotak, dapat diketahui perbedaan nilai siswa pada tiap kategori study time.

```python
Fjob_counts = df["Fjob"].value_counts()

plt.figure(figsize=(6,6))

plt.pie(Fjob_counts,
        labels=Fjob_counts.index,
        autopct="%1.1f%%",
        colors=["lightblue","orange","green","pink","yellow"])

plt.title("Distribusi Pekerjaan Ayah")

plt.show()
```

value_counts() digunakan untuk menghitung jumlah setiap kategori pekerjaan ayah. plt.pie() digunakan untuk membuat pie chart. labels menampilkan nama kategori. autopct menampilkan persentase setiap bagian pie chart. colors menentukan warna setiap bagian grafik.

Pie chart digunakan untuk menunjukkan persentase jenis pekerjaan ayah siswa. Dilihat dari ukuran setiap bagian lingkaran, dapat diketahui pekerjaan ayah yang paling banyak dalam dataset.

```python
explode = [0.1,0,0,0,0]
Fjob_counts = df["Fjob"].value_counts()
plt.figure(figsize=(6,6))

plt.pie(Fjob_counts,
        labels=Fjob_counts.index,
        autopct="%1.1f%%",
        explode=explode,
        colors=["cyan","pink","yellow","orange","lightgreen"])

plt.title("Exploded Pie Chart Pekerjaan Ayah")

plt.show()
```

explode digunakan untuk memisahkan salah satu bagian pie chart. Nilai 0.1 membuat bagian pertama sedikit keluar dari lingkaran. plt.pie() membuat grafik pie chart dengan efek exploded.

Exploded pie chart digunakan untuk menonjolkan salah satu kategori pekerjaan ayah sehingga bagian tersebut lebih mudah diperhatikan pada grafik.

```python
plt.figure(figsize=(10,6))

corr = df.corr(numeric_only=True)

sns.heatmap(corr,
            annot=True,
            cmap="coolwarm")

plt.title("Heatmap Korelasi Data")

plt.show()
```

Heatmap digunakan untuk melihat hubungan atau korelasi antar variabel dalam dataset. Pada kode di atas, pertama dilakukan perhitungan korelasi antar kolom numerik menggunakan df.corr(numeric_only=True) yang menghasilkan matriks korelasi. Nilai korelasi ini kemudian divisualisasikan menggunakan fungsi sns.heatmap() dari library Seaborn. Parameter annot=True digunakan untuk menampilkan nilai korelasi pada setiap kotak, cmap="coolwarm" digunakan untuk memberikan gradasi warna dari biru ke merah, dan linewidths=0.5 digunakan untuk memberi garis pembatas antar kotak. Dari heatmap ini dapat dilihat seberapa kuat hubungan antar variabel dalam dataset. Nilai korelasi yang mendekati 1 menunjukkan hubungan positif yang kuat, nilai mendekati -1 menunjukkan hubungan negatif yang kuat, sedangkan nilai yang mendekati 0 menunjukkan hubungan yang lemah atau tidak ada hubungan.
